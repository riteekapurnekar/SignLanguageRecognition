```python
#imprting modules
#reading the images input so that it can be trained
#Assertion error may be caused if in the gesture folder the id dont cosist of 1200 images
#ex: Gesture -> 0 -> should have 1200 images; Gesture ->1 have 50 images then error may occur
import cv2, os

def flip_images():
	gest_folder = "gestures"
	images_labels = []
	images = []
	labels = []
	for g_id in os.listdir(gest_folder):
		for i in range(1200):
			path = gest_folder+"/"+g_id+"/"+str(i+1)+".jpg"
			new_path = gest_folder+"/"+g_id+"/"+str(i+1+1200)+".jpg"
			print(path)
			img = cv2.imread(path, 0)
			img = cv2.flip(img, 1)
			cv2.imwrite(new_path, img)

flip_images()
```
